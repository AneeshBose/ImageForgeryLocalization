```
import random
def results():
  testpath = 'Data/Test/'
  c = 1
  with open('predictons.csv','w') as fp:
    fp.write('A,B,C\n')
    while c <= 50:
      strobj1=testpath+'image_id'+str(c)
      strobj='Image_id'+str(c)
      classify=int(random.getrandbits(1))#classifier(strobj1)
      strobj+=','+str(classify)+','
      if classify==1:
        strobj+='Mask/image_id'+str(c)+'.png'
        #getmask(strobj1)
      strobj+='\n'
      fp.write(strobj)
      c+=1
  fp.close()
```
